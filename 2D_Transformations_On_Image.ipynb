```python
#1) Translation P(x,y)------->P(x`,y`) shifting from one point to other point. tx->x cordinate shift, ty->y cordinate shift
# x`=x+tx   y`=y+ty
# the below is a matrix
#[x`]=[1,0][x]+[tx]
#[y`]=[0,1][y]+[ty]

#2) Rotation @-->theta angle
# x`=xcos@-ysin@  y`=xsin@-ycos@
# the below is a matrix
#[x`]=[cos@,-sin@][x]
#[y`]=[sin@,cos@][y]

#3) Scaling
#4) Shearing
#5) Reflection
```

```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("pauch.jpg")
print("Image Shape:",{img.shape})
print("Image Array:")
print(img)
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Loaded Image")
plt.show()
```

```python
img = cv2.imread("pauch.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# Translation
tx,ty = 400,300
translated_Matrix = np.float32([[1, 0, tx], [0, 1, ty]])
print(translated_Matrix)
translated = cv2.warpAffine(img, translated_Matrix, (img.shape[1], img.shape[0]))
```

```python
plt.figure(figsize=(6,6))
plt.imshow(translated)
plt.axis('off')
plt.title("Translated Image")
plt.show()
```

```python
img = cv2.imread("pauch.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# Rotation
angle = 60 #clock wise
center = (img.shape[1]//2,img.shape[0]//2)
rotation_matrix = cv2.getRotationMatrix2D(center,angle,1)
print(rotation_matrix)
rotated = cv2.warpAffine(img,rotation_matrix,(img.shape[1],img.shape[0]))
```

```python
plt.figure(figsize=(6,6))
plt.imshow(rotated)
plt.axis('off')
plt.title("Rotated Image")
plt.show()
```

```python
img = cv2.imread("pauch.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# Rotation
angle = -60 #Anticlock wise
center = (img.shape[1]//2,img.shape[0]//2)
rotation_matrix = cv2.getRotationMatrix2D(center,angle,1)
print(rotation_matrix)
rotated = cv2.warpAffine(img,rotation_matrix,(img.shape[1],img.shape[0]))
```

```python
plt.figure(figsize=(6,6))
plt.imshow(rotated)
plt.axis('off')
plt.title("Rotated Image")
plt.show()
```

```python
img = cv2.imread("pauch.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# Translation
tx,ty = 400,300
translated_Matrix = np.float32([[1, 0, tx], [0, 1, ty]])
print(translated_Matrix)
translated = cv2.warpAffine(img, translated_Matrix, (img.shape[1], img.shape[0]))

# Rotation
angle = 60 #clock wise
center = (translated.shape[1]//2,translated.shape[0]//2)
rotation_matrix = cv2.getRotationMatrix2D(center,angle,1)
print(rotation_matrix)
rotated = cv2.warpAffine(translated,rotation_matrix,(translated.shape[1],translated.shape[0]))
```

```python
plt.figure(figsize=(6,6))
plt.imshow(rotated)
plt.axis('off')
plt.title("Translated and Rotated Image")
plt.show()
```

```python
# Scaling
img = cv2.imread("pauch.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

scaled = cv2.resize(img,None,fx=4,fy=4,interpolation=cv2.INTER_LINEAR)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(scaled)
plt.axis('off')
plt.title("Scaled Image")
plt.show()
```

```python
# Shearing
img = cv2.imread("pauch.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

rows, cols, _ = img.shape
shear_matrix=np.float32([[1,0.2,0],[0.2,1,0]])
sheared=cv2.warpAffine(img,shear_matrix,(cols,rows))
```

```python
plt.figure(figsize=(6,6))
plt.imshow(sheared)
plt.axis('off')
plt.title("Sheared Image")
plt.show()
```

```python
# Reflection
img = cv2.imread("pauch.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

rows, cols, _ = img.shape
reflect_matrix=np.float32([[-1,0,cols],[0,1,0]])
reflected = cv2.warpAffine(img, reflect_matrix, (cols, rows))
```

```python
plt.figure(figsize=(6,6))
plt.imshow(reflected)
plt.axis('off')
plt.title("Reflected Image")
plt.show()
```
